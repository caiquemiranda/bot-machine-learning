```python
import pandas_ta as ta
```

##### Médias

- mme 7
- mme 9
- mme 21
- mme 50
- mme 66

```python
vale['sma7'] = ta.sma(vale['Adj Close'], length=7)
vale['sma9'] = ta.sma(vale['Adj Close'], length=9)
vale['sma21'] = ta.sma(vale['Adj Close'], length=21)
vale['sma50'] = ta.sma(vale['Adj Close'], length=50)
vale['sma66'] = ta.sma(vale['Adj Close'], length=66)
```

```python
vale.dropna(inplace=True)
vale.head()
```

```python
vale.info()
```

##### Distância médias

```python
vale['dist7'] = vale['Adj Close']-vale['sma7']
vale['dist9'] = vale['Adj Close']-vale['sma9']
vale['dist21'] = vale['Adj Close']-vale['sma21']
vale['dist50'] = vale['Adj Close']-vale['sma50']
vale['dist66'] = vale['Adj Close']-vale['sma66']
```

```python
vale.head()
```

```python
vale.info()
```

##### OBV

```python
# calculo OBV (on balance volume)
OBV = []
OBV.append(0)
for i in range(1, len(vale['Adj Close'])):
    if vale['Adj Close'][i] > vale['Adj Close'][i-1]: 
          OBV.append(OBV[-1] + vale['Volume'][i]) 
    elif vale['Adj Close'][i] < vale['Adj Close'][i-1]:
          OBV.append( OBV[-1] - vale['Volume'][i])
    else:
          OBV.append(OBV[-1])
```

```python
vale['obv'] = OBV
```

```python
vale.info()
```

##### Médias OBV

```python
vale['sma_obv_20'] = ta.sma(vale['obv'], length=20)
vale['sma_obv_50'] = ta.sma(vale['obv'], length=50)
vale['sma_obv_100'] = ta.sma(vale['obv'], length=100)
```
